```python
import sys, os
import numpy as np
from matplotlib import pyplot as plt
```

```python
parent_dir = os.path.abspath(os.pardir)
libpath = os.path.join(parent_dir, 'c3dp_source')
resultPath = os.path.join (parent_dir, 'results/misaligned_files')

if not libpath in sys.path:
    sys.path.insert(0, libpath)
```

```python
allignment_name1 = '43684.dat'
allignment_name2 = '43685.dat'
allignment_name3 = '43686.dat'
allignment_name4 = '43687.dat'
allignment_name5 = '43688.dat'
allignment_used_name = 'SNAP_43682_with_collimator.dat'
No_collimator_name = 'No_colli.dat'
```

```python
allignment1 =np.loadtxt( os.path.join (resultPath, allignment_name1), delimiter=",")
allignment2 =np.loadtxt( os.path.join (resultPath, allignment_name2), delimiter=",")
allignment3 =np.loadtxt( os.path.join (resultPath, allignment_name3), delimiter=",")
allignment4 =np.loadtxt( os.path.join (resultPath, allignment_name4), delimiter=",")
allignment5 =np.loadtxt( os.path.join (resultPath, allignment_name5), delimiter=",")
allignment_used =np.loadtxt( os.path.join (resultPath, allignment_used_name))
No_collimator =np.loadtxt( os.path.join (resultPath,No_collimator_name), delimiter=",")

d_1 = allignment1[:,0]
I_1 = allignment1[:,1]
e_1 = allignment1[:,2]

d_2 = allignment2[:,0]
I_2 = allignment2[:,1]
e_2 = allignment2[:,2]

d_3 = allignment3[:,0]
I_3 = allignment3[:,1]
e_3 = allignment3[:,2]

d_4 = allignment4[:,0]
I_4 = allignment4[:,1]
e_4 = allignment4[:,2]

d_5 = allignment5[:,0]
I_5 = allignment5[:,1]
e_5 = allignment5[:,2]

d_used = allignment_used[:,0]
I_used = allignment_used[:,1]
e_used = allignment_used[:,2]

d_noColli = No_collimator[:,0]
I_noColli = No_collimator[:,1]
e_noColli = No_collimator[:,2]
```

```python
%matplotlib inline
plt.figure(figsize=(15,7))
plt.errorbar (d_noColli, I_noColli, e_noColli, label ="no colli")

plt.plot (d_used, I_used-0.2, label ="withColli 1")
plt.plot (d_3, I_3, label ="with colli 2")

plt.legend()
# plt.errorbar (d_exp, I_exp, error_exp)
plt.xlim(3.0,3.5)
plt.ylim(1,2)
plt.xlabel('d_spacing ($\AA$)')
plt.ylabel('Intensity (arbitrary units)')
plt.show()
```
